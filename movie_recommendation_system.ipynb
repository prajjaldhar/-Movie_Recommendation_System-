```python
import ast
```

```python
#['Action', 'Adventure', 'Fantasy', 'Sci-Fi']
def format_dict(obj):
  genres=[]
  for i in ast.literal_eval(obj):
    genres.append(i['name'])
  return genres
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(format_dict)
```

```python
def character_format(obj):
    character_name=[]
    counter=0
    for i in ast.literal_eval(obj):
      if counter != 4:
        character_name.append(i['name'])
        counter+=1
      else:
         break
    return character_name
```

```python
movies['cast']=movies['cast'].apply(character_format)
```

```python
movies.head()
```

```python
movies.iloc[0].crew
```

```python
def fetch_director(obj):
  director_name=[]
  for i in ast.literal_eval(obj):
     if i['job']=='Director':
       director_name.append(i['name'])
       break
  return director_name
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem_words(text):
  words=[]
  for i in text.split():
    words.append(ps.stem(i))
  return " ".join(words)
```

```python
new_df['tags'].apply(stem_words)
```

we will count the distance between(the match between two movies that is defined using cosine angles)

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

we are doing sorting on the basis of 2nd data available that the distance and we are fetching the index value..... and grabbing the first 7 matched movies

```python
movies_list = sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])[1:7]
for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
def recommend(movie):
  try:
    movie_index=new_df[new_df['title'] == movie].index[0]
    distance_bw_movies = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance_bw_movies)),reverse=True,key = lambda x: x[1])[1:7]
    for i in movies_list:
      print(new_df.iloc[i[0]].title)
  except:
    print("Please try with correct Movie name")
```

```python
recommend("Avengers: Age of Ultron")
```
